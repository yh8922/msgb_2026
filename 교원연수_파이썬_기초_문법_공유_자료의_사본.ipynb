### 2. 시퀀스형 (Sequence Types)
여러 개의 원소를 순서대로 나열한 자료형입니다.
* **str**: 문자열(string)을 나타내며, 작은따옴표(')나 큰따옴표(")로 묶습니다. (예: `'Hello'`, `"Python"`)
* **list**: 리스트를 나타내며, `[]`로 묶습니다. 여러 자료형의 원소를 담을 수 있고, 수정이 가능합니다. (예: `[1, 'a', 3.14]`)
* **tuple**: 튜플을 나타내며, `()`로 묶습니다. 리스트와 비슷하지만, 한 번 정의된 원소는 수정할 수 없습니다. (예: `(1, 2, 'b')`)



````python
# 문자열 생성
my_string = "Hello, Python!"
````

````python
# 인덱싱 (특정 위치의 문자 접근)
print("첫 번째 문자:", my_string[0])  # H
print("마지막 문자:", my_string[-1]) # !

# 슬라이싱 (부분 문자열 추출)
print("2번째부터 5번째까지:", my_string[1:6]) # ello,
print("처음부터 7번째까지:", my_string[:7]) # Hello,
print("7번째부터 끝까지:", my_string[7:]) # Python!

# 문자열 길이
print("문자열 길이:", len(my_string)) # 14

# 문자열 연결
new_string = my_string + " Welcome!"
print(new_string) # Hello, Python! Welcome!
````

````python
# 문자열은 불변이므로 수정 시 오류 발생
my_string[0] = 'h' # TypeError: 'str' object does not support item assignment
````

````python
# 리스트 생성
my_list = [1, 'a', 3.14, True]

# 인덱싱
print("첫 번째 원소:", my_list[0])  # 1
print("마지막 원소:", my_list[-1]) # True

# 슬라이싱
print("1번째부터 3번째까지:", my_list[1:4]) # ['a', 3.14, True]

# 리스트 길이
print("리스트 길이:", len(my_list)) # 4

# 리스트 수정 (가변)
my_list[1] = 'new_value'
print("수정된 리스트:", my_list) # [1, 'new_value', 3.14, True]

# 원소 추가/삭제
my_list.append(10) # 리스트 끝에 추가
print("추가 후:", my_list) # [1, 'new_value', 3.14, True, 10]
my_list.remove('new_value') # 특정 원소 삭제
print("삭제 후:", my_list) # [1, 3.14, True, 10]
````

### 4. 집합형 (Set Types)
중복되지 않는 원소들의 모음입니다.
* **set**: 집합(set)을 나타내며, `{}`로 묶습니다. 중복된 원소는 자동으로 제거됩니다. (예: `{1, 2, 3, 2}`는 `{1, 2, 3}`이 됩니다)

* 비순차적 (Unordered): 항목이 저장된 순서가 없으므로 인덱싱(indexing)이나 슬라이싱(slicing)을 할 수 없습니다.

* 고유성 (Unique): 집합 내의 모든 요소는 고유해야 하며, 중복된 값을 추가하려고 해도 무시됩니다.

* 가변성 (Mutable): 일반적인 set 객체는 생성 후 요소를 추가하거나 제거할 수 있습니다.

* 요소 제약: 집합의 요소는 immutable (변경 불가능)한 자료형(예: int, str, tuple)만 가능합니다. list나 dict 같은 mutable 객체는 요소로 사용할 수 없습니다.

* 주요 용도: 중복 제거, 멤버십 테스트 (어떤 요소가 집합에 있는지 빠르게 확인), 수학적 집합 연산 (합집합, 교집합, 차집합)에 사용됩니다.

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# 합집합 (Union)
print(f"합집합: {A | B}")         # 출력: {1, 2, 3, 4, 5, 6}

# 교집합 (Intersection)
print(f"교집합: {A & B}")       # 출력: {3, 4}

# 차집합 (Difference)
print(f"차집합: {A - B}")         # 출력: {1, 2}
````

## 논리 연산자

두 개 이상의 논리 값 또는 조건을 결합하여 새로운 논리 값을 만드는 데 사용됩니다.

| 연산자 | 이름 | 설명 | 예시 |
| :--- | :--- | :--- | :--- |
| `and` | 논리곱 | 양쪽의 값이 모두 **True**일 때만 **True**를 반환합니다. | `True and False` $\\rightarrow$ `False` |
| `or` | 논리합 | 둘 중 **하나라도 True**이면 **True**를 반환합니다. | `True or False` $\\rightarrow$ `True` |
| `not` | 논리 부정 | 현재 논리 값을 **반전**시킵니다. | `not True` $\\rightarrow$ `False` |

-----
#### False로 간주되는 값 (Falsy Values)

프로그램의 흐름을 제어할 때 **거짓**으로 취급되어 `if` 구문 등을 건너뛰게 만드는 값들입니다.

  * **숫자**: `0`, `0.0` (모든 종류의 0)
  * **비어 있는 컬렉션**: `''` (빈 문자열), `[]` (빈 리스트), `{}` (빈 딕셔너리), `set()` (빈 집합), `()` (빈 튜플)
  * **특수 값**: `None`

#### True로 간주되는 값 (Truthy Values)

위의 Falsy 값들을 제외한 **나머지 모든 값**은 **참**으로 간주됩니다.

  * **숫자**: `1`, `-10`, `0.001` (0이 아닌 모든 수)
  * **채워진 컬렉션**: `'Hello'`, `[1, 2]`, `{'a': 1}` (요소가 하나라도 있는 모든 컬렉션)


### 자료형 변환(Data Type Casting)
파이썬에서는 int(), float(), str(), list(), tuple() 등의 내장 함수를 사용해 여러 자료형을 서로 변환할 수 있습니다.

#### **주요 자료형 변환 함수**
- int(x): 실수나 문자열을 정수형으로 변환합니다.
- float(x): 정수나 문자열을 실수형으로 변환합니다.
- str(x): 숫자(정수, 실수)나 다른 자료형을 문자열로 변환합니다.
- list(x): 반복 가능한 객체(문자열, 튜플 등)를 리스트로 변환합니다.
- tuple(x): 반복 가능한 객체(리스트, 문자열 등)를 튜플로 변환합니다.
- set(x): 반복 가능한 객체를 집합(set)으로 변환합니다.

### 파이썬 mutable과 immutable
Mutable (변경 가능) 객체:

* 생성된 후에도 내부 값을 변경할 수 있는 객체입니다.

* 메모리 주소는 그대로 유지하면서 내용만 바뀝니다.

> 예시: 리스트에 append()나 pop()을 사용해 항목을 추가하거나 삭제할 수 있습니다.

* 한번 생성되면 내용을 변경할 수 없는 객체입니다.

* 변경이 필요하면 기존 객체를 복사하여 새로운 객체를 생성합니다.

* 종류: int, float, str, tuple, frozenset 등이 있습니다.

> 예시: 문자열에 새로운 문자를 추가하면, 기존 문자열은 그대로 두고 새로운 문자열 객체가 생성됩니다.

## print 함수

print() 함수는 파이썬 프로그램에서 데이터를 화면(콘솔)에 출력하는 역할을 하는 내장 함수.<br>
- 출력: 개발자가 프로그램의 실행 결과를 확인하거나, 사용자에게 특정 정보를 보여주고자 할 때 사용합니다.
- 다양한 데이터 출력: 문자열, 숫자, 변수 등 모든 종류의 데이터를 인자로 받아 출력. 어떤 데이터를 넣든 내부적으로 문자열로 변환하여 출력합니다.
- 기본 동작: 여러 개의 인자를 전달하면 기본적으로 공백으로 구분해서 출력하며, 출력이 끝난 후에는 자동으로 줄 바꿈을 합니다. 이 동작은 sep나 end 같은 매개변수로 변경할 수 있습니다.

파이썬의 내장 함수 print()의 함수 원형(signature)은 다음과 같습니다.
> `print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)`

<br>각 매개변수의 의미는 다음과 같습니다.
- *objects: 출력할 하나 이상의 객체들. *가 붙어 있어 쉼표로 구분된 여러 개의 값을 전달할 수 있습니다.
- sep=' ': 여러 개의 객체를 출력할 때 객체들 사이에 삽입할 구분 문자열. 기본값은 공백 문자(' ')입니다.
- end='\n': 출력의 마지막에 추가할 문자열. 기본값은 줄바꿈 문자('\n')입니다.
- file=sys.stdout: 출력할 대상을 지정합니다. 기본값은 sys.stdout으로, 표준 출력(콘솔)을 의미합니다. 파일을 지정하면 파일에 내용을 쓸 수 있습니다.
- flush=False: 버퍼를 강제로 비울지 여부를 결정합니다. 기본값은 False로, 성능을 위해 버퍼가 가득 차거나 줄바꿈 문자가 나올 때까지 기다립니다. True로 설정하면 즉시 출력합니다.

# 🎯조건문

**조건문**은 프로그램의 **흐름을 제어**하는 핵심 문법.
<br>특정 조건이 \*\*참(True)\*\*인지 \*\*거짓(False)\*\*인지에 따라 실행할 코드 블록을 결정합니다.<br>실생활에서 "만약 ~이라면, ~을 해라"라고 말하는 것과 같은 역할을 합니다.
## 조건문의 핵심 개념
- 결정하기: 프로그램이 다양한 상황이나 사용자의 입력에 따라 다르게 동작하도록 만듭니다.
- 조건식: 조건문은 '참(True)' 또는 '거짓(False)'으로 평가되는 논리식(Boolean expression)을 사용합니다.
- 코드 블록: 조건이 참일 때 실행할 코드 묶음을 정의하는데, 파이썬에서는 들여쓰기(indentation)로 이 코드 블록을 구분합니다.
- 키워드: `if`, `elif`, `else` 키워드를 사용하여 다양한 조건에 대한 논리를 구성합니다

### 기본 구조: `if`, `elif`, `else`

#### `if` (만약 ~라면)

가장 기본이 되는 조건문입니다. `if` 뒤의 조건이 참일 때만 해당 코드 블록을 실행합니다.

````python
score = 85
if score >= 60:
    print("합격입니다.")
````

#### `else` (그렇지 않다면)

`if`의 조건이 거짓일 때 실행되는 코드를 정의합니다. `else`는 `if` 없이 단독으로 사용할 수 없습니다.

````python
score = 45
if score >= 60:
    print("합격입니다.")
else:
    print("불합격입니다.")
````

#### `elif` (그렇지 않고 만약 ~라면)

여러 개의 조건을 순서대로 검사할 때 사용합니다. 앞선 `if`나 `elif`의 조건이 모두 거짓일 때, 현재 `elif`의 조건을 확인합니다.

````python
score = 75
if score >= 90:
    print("A 학점")
elif score >= 80:
    print("B 학점")
elif score >= 70:
    print("C 학점")
else:
    print("F 학점")
````

## 다양한 사용법

#### 논리 연산자와의 결합

조건식을 만들 때 **비교 연산자**(`==`, `!=`, `>`, `<`, `>=`, `<=`)와 **논리 연산자**(`and`, `or`, `not`)를 함께 사용해 복잡한 조건을 표현할 수 있습니다.


````python
age = 25
is_student = True

if age >= 18 and is_student:
    print("학생 할인이 적용됩니다.")
````

#### 조건부 표현식 (삼항 연산자)

코드를 한 줄로 간결하게 표현하고 싶을 때 유용합니다. 다른 언어의 삼항 연산자와 비슷한 역할을 합니다.

**형식:** `결과1 if 조건 else 결과2`



````python
temperature = 35

status = "더움" if temperature > 30 else "보통"
print(status)
````

#### 중첩 조건문

하나의 조건문 안에 또 다른 조건문을 넣어 더 세밀한 조건을 처리할 수 있습니다.

## 🔄 for문 (For Loop)
- 특정한 자료형(시퀀스)에 들어있는 요소를 하나씩 꺼내어 지정된 횟수만큼 반복적으로 코드를 실행하는 제어문.
- 프로그래밍에서 반복되는 작업을 효율적으로 처리하기 위해 사용.
- for 반복문의 개념
> 순회 가능한 객체(Iterable)
for 반복문은 문자열, 리스트, 튜플, 딕셔너리 등 순서가 있는(혹은 순서대로 꺼낼 수 있는) 객체인 '이터러블'을 순회합니다. 이 객체의 요소가 모두 소진될 때까지 반복문을 실행합니다.

### 기본 문법
```python
for 변수 in 반복가능객체:
    실행할 코드
```

### 리스트와 for문

````python
# 기본 예제
fruits = ['사과', '바나나', '오렌지']
for fruit in fruits:
    print(f"과일: {fruit}")
````

````python
# range() 함수 활용

# range(stop)
for i in range(5):
    print(i)  # 0, 1, 2, 3, 4

# range(start, stop)
for i in range(2, 7):
    print(i)  # 2, 3, 4, 5, 6

# range(start, stop, step)
for i in range(1, 10, 2):
    print(i)  # 1, 3, 5, 7, 9
````

````python
# 문자열과 for문

word = "Python"
for char in word:
    print(char)  # P, y, t, h, o, n
````

````python
# 딕셔너리와 for문

student = {'이름': '김철수', '나이': 20, '전공': '컴퓨터공학'}

# 키만 반복
for key in student:
    print(key)

# 값만 반복
for value in student.values():
    print(value)

# 키-값 쌍 반복
for key, value in student.items():
    print(f"{key}: {value}")
````

````python
### enumerate() 함수

fruits = ['사과', '바나나', '오렌지']

# 인덱스와 값을 함께 사용
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# 시작 인덱스 지정
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}번째: {fruit}")

````

## 🔁 while문 (While Loop)
while 반복문은 특정 조건식이 참(True)인 동안 코드 블록을 계속 반복해서 실행하는 제어문.
- for 반복문이 정해진 횟수만큼 반복하는 데 주로 사용되는 반면, while 반복문은 반복 횟수가 불확실할 때, 즉 조건이 거짓(False)이 될 때까지 반복해야 할 때 사용.
- while 반복문의 핵심 개념

> 1. 조건식 : while문은 시작하기 전에 먼저 조건식을 평가합니다.
조건식이 참(True)이면 코드 블록을 실행하고, 다시 조건식으로 돌아가 재평가합니다.
이 과정은 조건식이 거짓(False)이 될 때까지 계속 반복됩니다.
2. 무한 루프와 종료 조건
만약 while문의 조건식이 항상 참인 상태로 유지되면, 반복문은 영원히 끝나지 않는 **무한 루프(infinite loop)**에 빠지게 됩니다.
이를 방지하려면 반복문 내부에서 조건식에 영향을 미치는 변수의 값을 변경하여 언젠가 조건이 거짓이 되도록 만들어야 합니다.

### 기본 문법
```python
while 조건:
    실행할 코드
```

# 📑함수


## 함수의 정의와 개념

## 함수란?

함수는 특정 작업을 수행하는 코드 블록으로, 재사용 가능하고 모듈화된 프로그램을 작성하는 핵심 요소입니다. 한 번 정의하면 필요할 때마다 호출하여 사용할 수 있습니다.

## 기본 함수 정의

```python
def 함수명(매개변수):
    """독스트링(선택사항)"""
    # 실행할 코드
    return 반환값
```


````python
# 간단한 예제

def greet(name):
    """이름을 받아 인사하는 함수"""
    return f"안녕하세요, {name}님!"

result = greet("철수")
print(result)  # 안녕하세요, 철수님!
````

## 다양한 함수 사용법

````python
# 매개변수가 없는 함수
def say_hello():
    print("Hello, World!")

say_hello()  # Hello, World!
````

````python
# 여러 매개변수를 받는 함수

def add(a, b):
    return a + b

result = add(5, 3)  # 8
````

### 가변 인자 (*args)

위치 인자를 여러 개 받을 수 있습니다.

````python
def sum_all(*numbers):
    return sum(numbers)

print(sum_all(1, 2, 3))  # 6
print(sum_all(1, 2, 3, 4, 5))  # 15
````

````python
# 키워드 가변 인자 (**kwargs)

# 키-값 쌍의 인자를 여러 개 받을 수 있습니다.

def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="김철수", age=25, city="서울")
````

````python
# 위치 인자와 키워드 인자 혼용

def introduce(name, age, *hobbies, **details):
    print(f"이름: {name}, 나이: {age}")
    print(f"취미: {', '.join(hobbies)}")
    print(f"추가 정보: {details}")

introduce("홍길동", 30, "독서", "운동", city="부산", job="개발자")
````

````python
# 여러 값 반환
def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

minimum, maximum, average = get_stats([1, 2, 3, 4, 5])
print(f"최소: {minimum}, 최대: {maximum}, 평균: {average}")
````

### 람다 함수 (익명 함수)

### 재귀 함수

재귀 함수는 자기 자신을 호출하는 함수입니다. 반복문 대신 사용할 수 있으며, 특정 문제를 더 직관적으로 해결할 수 있습니다.

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # 120
# 실행 과정:
# factorial(5) = 5 * factorial(4)
# factorial(4) = 4 * factorial(3)
# factorial(3) = 3 * factorial(2)
# factorial(2) = 2 * factorial(1)
# factorial(1) = 1
# 결과: 5 * 4 * 3 * 2 * 1 = 120
```

### 중첩 함수 (내부 함수)
함수 내부에 다른 함수를 정의하는 것을 중첩 함수.

헬퍼 함수를 외부에 노출하지 않아 캡슐화가 가능.
```python
def outer_function(text):
    def inner_function():
        print(text)
    
    inner_function()

outer_function("안녕하세요")  # 안녕하세요
```

### 클로저
클로저는 중첩 함수가 외부 함수의 변수를 "기억"하는 기능입니다.

외부 함수가 종료된 후에도 내부 함수가 외부 함수의 변수에 접근할 수 있습니다.

클로저의 특징:

- 데이터 은닉: 외부에서 직접 접근할 수 없는 변수를 가질 수 있습니다
- 상태 유지: 함수 호출 사이에 상태를 유지할 수 있습니다
- 팩토리 패턴: 비슷하지만 다른 동작을 하는 여러 함수를 생성할 수 있습니다
```python
def multiplier(n):
    def multiply(x):
        return x * n
    return multiply

times_three = multiplier(3)
print(times_three(10))  # 30
```

### 데코레이터

기존 함수를 수정하지 않고 기능을 추가할 수 있습니다.

@ 기호를 사용하여 적용합니다.
데코레이터의 장점:

- 코드 재사용성: 공통 기능을 여러 함수에 쉽게 적용할 수 있습니다
- 가독성: 함수의 의도를 명확하게 표현할 수 있습니다
- 관심사 분리: 핵심 로직과 부가 기능을 분리할 수 있습니다
```python
def uppercase_decorator(func):
    """함수의 반환값을 대문자로 변환하는 데코레이터"""
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

@uppercase_decorator
def greet():
    return "hello, world"

print(greet())  # HELLO, WORLD

# 위 코드는 아래와 동일합니다:
# def greet():
#     return "hello, world"
# greet = uppercase_decorator(greet)
```

### 타입 힌트 (Type Hints)

Python 3.5 이상에서 함수의 매개변수와 반환값 타입을 명시할 수 있습니다.

```python
def add(a: int, b: int) -> int:
    return a + b

def greet(name: str) -> str:
    return f"Hello, {name}!"
```

## 함수 사용의 장점

**코드 재사용성**: 같은 코드를 반복해서 작성할 필요가 없습니다.

**모듈화**: 복잡한 프로그램을 작은 단위로 나누어 관리할 수 있습니다.

**가독성 향상**: 코드의 의도를 명확하게 표현할 수 있습니다.

**유지보수 용이**: 한 곳에서 수정하면 해당 함수를 사용하는 모든 곳에 적용됩니다.

**테스트 용이**: 독립적인 단위로 테스트할 수 있습니다.

# 📚 클래스

## 클래스란?

클래스(Class)는 객체를 만들기 위한 설계도 또는 틀입니다. 관련된 데이터(속성)와 기능(메서드)을 하나로 묶어서 관리할 수 있게 해주는 객체 지향 프로그래밍의 핵심 개념입니다.

### 기본 용어
- **클래스(Class)**: 객체를 만들기 위한 설계도
- **객체(Object)**: 클래스로부터 만들어진 실체 (인스턴스라고도 함)
- **속성(Attribute)**: 객체가 가지는 데이터 (변수)
- **메서드(Method)**: 객체가 수행할 수 있는 동작 (함수)

---

## 1. 기본 클래스 정의

 **도서관 시스템**

````python
# 기본 클래스 정의
class Book:
    """책을 나타내는 클래스"""
    pass  # 아무 내용도 없는 클래스

# 객체 생성 (인스턴스화)
book1 = Book()
book2 = Book()

print(type(book1))  # <class '__main__.Book'>
print(book1 == book2)  # False (서로 다른 객체)
````

````python
# 속성(Attribute) 추가
# 인스턴스 속성

class Book:
    """책을 나타내는 클래스"""

    def __init__(self, title, author, isbn, year):
        """
        생성자 메서드: 객체가 생성될 때 자동으로 호출됨

        Parameters:
        - self: 객체 자신을 가리키는 참조
        - title: 책 제목
        - author: 저자
        - isbn: ISBN 번호
        - year: 출판년도
        """
        self.title = title      # 인스턴스 속성
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True  # 대출 가능 여부 (기본값)

# 객체 생성
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
book2 = Book("데이터 분석 입문", "이영희", "978-89-01234-57-8", 2023)

# 속성 접근
print(book1.title)       # 파이썬 프로그래밍
print(book1.author)      # 김철수
print(book1.is_available)  # True

# 속성 수정
book1.is_available = False
print(book1.is_available)  # False
````

````python
# 클래스 속성


class Book:
    """책을 나타내는 클래스"""

    # 클래스 속성: 모든 인스턴스가 공유
    library_name = "시립 중앙도서관"
    total_books = 0

    def __init__(self, title, author, isbn, year):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True

        # 책이 생성될 때마다 카운트 증가
        Book.total_books += 1

# 객체 생성
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
book2 = Book("데이터 분석 입문", "이영희", "978-89-01234-57-8", 2023)
book3 = Book("웹 개발 기초", "박민수", "978-89-01234-58-9", 2024)

# 클래스 속성 접근
print(Book.library_name)  # 시립 중앙도서관
print(Book.total_books)   # 3

# 인스턴스를 통해서도 접근 가능
print(book1.library_name)  # 시립 중앙도서관
````

````python
# 메서드(Method) 추가

class Book:
    """책을 나타내는 클래스"""

    library_name = "시립 중앙도서관"
    total_books = 0

    def __init__(self, title, author, isbn, year):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True
        self.borrowed_by = None
        Book.total_books += 1

    # 인스턴스 메서드
    def get_info(self):
        """책 정보를 문자열로 반환"""
        status = "대출 가능" if self.is_available else f"대출 중 ({self.borrowed_by})"
        return f"'{self.title}' - {self.author} ({self.year}) [{status}]"

    def borrow(self, borrower_name):
        """책을 대출하는 메서드"""
        if self.is_available:
            self.is_available = False
            self.borrowed_by = borrower_name
            return f"'{self.title}'이(가) {borrower_name}님에게 대출되었습니다."
        else:
            return f"'{self.title}'은(는) 현재 {self.borrowed_by}님이 대출 중입니다."

    def return_book(self):
        """책을 반납하는 메서드"""
        if not self.is_available:
            borrower = self.borrowed_by
            self.is_available = True
            self.borrowed_by = None
            return f"'{self.title}'이(가) {borrower}님에 의해 반납되었습니다."
        else:
            return f"'{self.title}'은(는) 이미 도서관에 있습니다."

    def is_classic(self):
        """출판된 지 20년이 넘었는지 확인"""
        from datetime import datetime
        current_year = datetime.now().year
        return current_year - self.year > 20

# 사용 예제
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
book2 = Book("클린 코드", "로버트 마틴", "978-89-01234-59-0", 2000)

# 메서드 호출
print(book1.get_info())
# '파이썬 프로그래밍' - 김철수 (2024) [대출 가능]

print(book1.borrow("홍길동"))
# '파이썬 프로그래밍'이(가) 홍길동님에게 대출되었습니다.

print(book1.get_info())
# '파이썬 프로그래밍' - 김철수 (2024) [대출 중 (홍길동)]

print(book1.borrow("김영희"))
# '파이썬 프로그래밍'은(는) 현재 홍길동님이 대출 중입니다.

print(book1.return_book())
# '파이썬 프로그래밍'이(가) 홍길동님에 의해 반납되었습니다.

print(book2.is_classic())  # True (2000년 출판)
print(book1.is_classic())  # False (2024년 출판)
````

## 4. 특수 메서드 (Magic Methods)

특수 메서드는 `__메서드명__` 형태로 작성하며, 특정 상황에서 자동으로 호출됩니다.

매직 메서드는 파이썬 클래스 내에서 특별한 의미를 가지는 메서드입니다.

이름 앞뒤에 더블 언더스코어(__)가 붙어 있어서 "던더(Dunder) 메서드"라고도 불립니다.

````python
class Book:
    """책을 나타내는 클래스"""

    library_name = "시립 중앙도서관"
    total_books = 0

    def __init__(self, title, author, isbn, year):
        """
        언제 호출되나요? book1 = Book(...)처럼 클래스의 새 객체(인스턴스)를 만들 때 가장 먼저 호출.
        무엇을 하나요? 객체가 생성될 때 필요한 초기 속성들(책 제목, 저자 등)을 설정."""
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True
        self.borrowed_by = None
        Book.total_books += 1

    def __str__(self):
        """
        사용자용 문자열 표현
        언제 호출되나요? print(book1) 또는 str(book1)을 사용할 때 호출됩니다.
        무엇을 하나요? 객체를 사람이 보기 좋은, 이해하기 쉬운 문자열로 표현합니다.
        """
        status = "대출 가능" if self.is_available else f"대출 중"
        return f"'{self.title}' by {self.author} [{status}]"

    def __repr__(self):
        """
        공식 문자열 표현: 개발자를 위한 출력
        객체를 재생성할 수 있는 정보 제공

        언제 호출되나요? repr(book1)을 사용하거나, 인터랙티브 콘솔에서 객체 변수명만 입력했을 때 호출됩니다. __str__이 정의되지 않았다면 print()에서도 호출됩니다.
        무엇을 하나요? 객체를 명확하게 식별하거나, 심지어 그 객체를 다시 생성할 수 있는 형태의 문자열로 표현하는 것을 목표로 합니다.
        """
        return f"Book('{self.title}', '{self.author}', '{self.isbn}', {self.year})"

    def __len__(self):
        """len() 함수 호출 시 동작 정의"""
        # 책 제목의 길이 반환
        return len(self.title)

    def __eq__(self, other):
        """
        동등성 비교

        언제 호출되나요? book1 == book3 처럼 == 연산자를 사용할 때 호출됩니다.
        무엇을 하나요? 두 객체가 '같은지'를 어떤 기준으로 판단할지 정의합니다. Book 클래스에서는 ISBN 번호가 같으면 같은 책으로 판단하도록 정의했습니다. 여기서 isinstance(other, Book)를 사용해 비교 대상이 Book 객체가 맞는지 먼저 확인하는 것이 안전한 코드를 작성하는 좋은 방법입니다.
        """
        if isinstance(other, Book):
            return self.isbn == other.isbn  # ISBN이 같으면 같은 책
        return False

    def __lt__(self, other):
        """대소 비교 (Less Than)"""
        if isinstance(other, Book):
            return self.year < other.year  # 출판년도로 비교
        return NotImplemented

    def __del__(self):
        """소멸자: 객체가 삭제될 때 호출
        언제 호출되나요? 객체가 메모리에서 완전히 사라질 때(참조 카운트가 0이 될 때) 호출됩니다.
        무엇을 하나요? 객체가 소멸되기 직전에 마무리 작업을 수행합니다. 예제에서는 전체 책의 수를 하나 줄이는 역할을 합니다.
        """
        Book.total_books -= 1
        print(f"'{self.title}' 객체가 삭제되었습니다.")

# 사용 예제
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
book2 = Book("데이터 분석 입문", "이영희", "978-89-01234-57-8", 2023)
book3 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)

# __str__ 사용
print(book1)  # '파이썬 프로그래밍' by 김철수 [대출 가능]

# __repr__ 사용
print(repr(book1))
# Book('파이썬 프로그래밍', '김철수', '978-89-01234-56-7', 2024)

# __len__ 사용
print(len(book1))  # 10 (파이썬 프로그래밍의 글자 수)

# __eq__ 사용
print(book1 == book3)  # True (같은 ISBN)
print(book1 == book2)  # False (다른 ISBN)

# __lt__ 사용 (정렬)
books = [book1, book2]
sorted_books = sorted(books)  # 출판년도 순으로 정렬
for book in sorted_books:
    print(book)
# '데이터 분석 입문' by 이영희 [대출 가능]
# '파이썬 프로그래밍' by 김철수 [대출 가능]
````

````python
# 클래스 메서드와 정적 메서드

class Book:
    """책을 나타내는 클래스"""

    library_name = "시립 중앙도서관"
    total_books = 0

    def __init__(self, title, author, isbn, year):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True
        self.borrowed_by = None
        Book.total_books += 1

    def __str__(self):
        status = "대출 가능" if self.is_available else f"대출 중"
        return f"'{self.title}' by {self.author} [{status}]"

    # 인스턴스 메서드 (self 사용)
    def borrow(self, borrower_name):
        """책을 대출하는 메서드"""
        if self.is_available:
            self.is_available = False
            self.borrowed_by = borrower_name
            return f"'{self.title}'이(가) {borrower_name}님에게 대출되었습니다."
        else:
            return f"'{self.title}'은(는) 현재 대출 중입니다."

    # 클래스 메서드 (cls 사용)
    @classmethod
    def get_total_books(cls):
        """전체 책 개수를 반환하는 클래스 메서드"""
        return f"{cls.library_name}의 총 장서 수: {cls.total_books}권"

    @classmethod
    def change_library_name(cls, new_name):
        """도서관 이름을 변경하는 클래스 메서드"""
        old_name = cls.library_name
        cls.library_name = new_name
        return f"도서관 이름이 '{old_name}'에서 '{new_name}'으로 변경되었습니다."

    @classmethod
    def from_string(cls, book_string):
        """
        문자열에서 Book 객체를 생성하는 팩토리 메서드
        형식: "제목|저자|ISBN|년도"
        """
        title, author, isbn, year = book_string.split('|')
        return cls(title, author, isbn, int(year))

    # 정적 메서드 (self, cls 둘 다 사용 안 함)
    @staticmethod
    def is_valid_isbn(isbn):
        """ISBN 형식이 유효한지 검증하는 정적 메서드"""
        # 간단한 검증: "978-"로 시작하고 충분한 길이인지
        isbn = isbn.replace('-', '')
        return isbn.startswith('978') and len(isbn) == 13 and isbn.isdigit()

    @staticmethod
    def calculate_late_fee(days_late):
        """연체료를 계산하는 정적 메서드"""
        fee_per_day = 100  # 일일 연체료 100원
        return days_late * fee_per_day

# 사용 예제

# 1) 인스턴스 메서드 사용
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
print(book1.borrow("홍길동"))
# '파이썬 프로그래밍'이(가) 홍길동님에게 대출되었습니다.

# 2) 클래스 메서드 사용
book2 = Book("데이터 분석 입문", "이영희", "978-89-01234-57-8", 2023)
print(Book.get_total_books())
# 시립 중앙도서관의 총 장서 수: 2권

print(Book.change_library_name("국립 중앙도서관"))
# 도서관 이름이 '시립 중앙도서관'에서 '국립 중앙도서관'으로 변경되었습니다.

# 팩토리 메서드로 객체 생성
book3 = Book.from_string("웹 개발 기초|박민수|978-89-01234-58-9|2024")
print(book3)  # '웹 개발 기초' by 박민수 [대출 가능]

# 3) 정적 메서드 사용
print(Book.is_valid_isbn("978-89-01234-56-7"))  # True
print(Book.is_valid_isbn("123-45-67890"))       # False

print(f"5일 연체료: {Book.calculate_late_fee(5)}원")  # 5일 연체료: 500원
print(f"10일 연체료: {Book.calculate_late_fee(10)}원")  # 10일 연체료: 1000원

# 인스턴스를 통해서도 호출 가능
print(book1.is_valid_isbn("978-89-01234-56-7"))  # True
````

**차이점 정리:**
- **인스턴스 메서드**: 특정 객체의 데이터를 사용/수정 (`self` 사용)
- **클래스 메서드**: 클래스 전체와 관련된 작업 (`cls` 사용)
- **정적 메서드**: 클래스와 관련은 있지만 클래스/인스턴스 데이터를 사용하지 않음

# 상속(Inheritance)

상속은 기존 클래스(부모 클래스)의 모든 속성과 메서드를 물려받아 새로운 클래스(자식 클래스)를 만드는 것을 의미.
- 부모 클래스 (Parent Class / Superclass): 속성과 메서드를 물려주는 클래스. (예: Book)
- 자식 클래스 (Child Class / Subclass): 속성과 메서드를 물려받는 클래스. (예: EBook)

이를 통해 자식 클래스는 부모 클래스의 기능을 그대로 사용하면서, 자신만의 새로운 기능을 추가하거나 기존 기능을 변경(오버라이딩)할 수 있습니다.

````python
class Book:
    """기본 책 클래스 (부모 클래스)"""

    library_name = "시립 중앙도서관"
    total_books = 0

    def __init__(self, title, author, isbn, year):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year = year
        self.is_available = True
        self.borrowed_by = None
        Book.total_books += 1

    def __str__(self):
        status = "대출 가능" if self.is_available else f"대출 중"
        return f"'{self.title}' by {self.author} [{status}]"

    def borrow(self, borrower_name):
        """책을 대출하는 메서드"""
        if self.is_available:
            self.is_available = False
            self.borrowed_by = borrower_name
            return f"'{self.title}'이(가) {borrower_name}님에게 대출되었습니다."
        else:
            return f"'{self.title}'은(는) 현재 대출 중입니다."

    def return_book(self):
        """책을 반납하는 메서드"""
        if not self.is_available:
            borrower = self.borrowed_by
            self.is_available = True
            self.borrowed_by = None
            return f"'{self.title}'이(가) 반납되었습니다."
        else:
            return f"'{self.title}'은(는) 이미 도서관에 있습니다."

    def get_info(self):
        """책 정보 반환"""
        return f"제목: {self.title}\n저자: {self.author}\nISBN: {self.isbn}\n출판년도: {self.year}"


# 자식 클래스 1: 전자책
class EBook(Book):
    """전자책 클래스 (Book을 상속)"""

    def __init__(self, title, author, isbn, year, file_size, format):
        # 부모 클래스의 __init__ 호출
        super().__init__(title, author, isbn, year)
        # 전자책만의 속성 추가
        self.file_size = file_size  # MB 단위
        self.format = format        # PDF, EPUB 등
        self.download_count = 0

    def __str__(self):
        # 부모의 __str__을 확장
        base_str = super().__str__()
        return f"{base_str} [전자책: {self.format}, {self.file_size}MB]"

    def download(self, user_name):
        """전자책 다운로드 메서드"""
        self.download_count += 1
        return f"'{self.title}'이(가) {user_name}님에게 다운로드되었습니다. (총 {self.download_count}회)"

    def get_info(self):
        """전자책 정보 반환 (메서드 오버라이딩)"""
        base_info = super().get_info()
        return f"{base_info}\n파일 형식: {self.format}\n파일 크기: {self.file_size}MB\n다운로드 횟수: {self.download_count}회"


# 자식 클래스 2: 참고서
class ReferenceBook(Book):
    """참고서 클래스 (Book을 상속)"""

    def __init__(self, title, author, isbn, year, category, edition):
        super().__init__(title, author, isbn, year)
        self.category = category  # 분야 (역사, 과학, 철학 등)
        self.edition = edition    # 판 (1판, 2판 등)
        self.can_borrow = False   # 참고서는 대출 불가

    def __str__(self):
        return f"[참고서] '{self.title}' by {self.author} ({self.category}, {self.edition}판) [관내 열람만 가능]"

    def borrow(self, borrower_name):
        """참고서는 대출 불가 (메서드 오버라이딩)"""
        return f"'{self.title}'은(는) 참고서로 대출이 불가능합니다. 관내에서만 열람 가능합니다."

    def get_info(self):
        """참고서 정보 반환"""
        base_info = super().get_info()
        return f"{base_info}\n분야: {self.category}\n판: {self.edition}판\n대출 가능: 불가 (관내 열람만)"


# 사용 예제

# 1) 일반 책
print("=== 일반 책 ===")
book1 = Book("파이썬 프로그래밍", "김철수", "978-89-01234-56-7", 2024)
print(book1)
print(book1.borrow("홍길동"))
print()

# 2) 전자책
print("=== 전자책 ===")
ebook1 = EBook("클린 코드", "로버트 마틴", "978-89-01234-60-4", 2008, 15.5, "PDF")
print(ebook1)
print(ebook1.download("김영희"))
print(ebook1.download("이철수"))
print(ebook1.get_info())
print()

# 3) 참고서
print("=== 참고서 ===")
ref_book1 = ReferenceBook("한국사 대사전", "역사학회", "978-89-01234-61-1", 2020, "역사", 3)
print(ref_book1)
print(ref_book1.borrow("박민수"))  # 대출 불가 메시지
print(ref_book1.get_info())
print()

# 4) isinstance()와 issubclass() 확인
print("=== 타입 확인 ===")
print(f"ebook1은 EBook의 인스턴스? {isinstance(ebook1, EBook)}")      # True
print(f"ebook1은 Book의 인스턴스? {isinstance(ebook1, Book)}")          # True
print(f"book1은 EBook의 인스턴스? {isinstance(book1, EBook)}")          # False
print(f"EBook은 Book의 서브클래스? {issubclass(EBook, Book)}")         # True
print(f"ReferenceBook은 Book의 서브클래스? {issubclass(ReferenceBook, Book)}")  # True
````

# 🐍 파이썬 예외 처리 (Exception Handling)
자료예외 처리는 프로그램의 **견고성(Robustness)**을 확보하고 예기치 않은 종료를 방지하기 위한 핵심 기술입니다.



## 예외(Exception)와 오류(Error)
### 예외란 무엇인가?
**예외(Exception)** 는 프로그램 실행 중에 발생하는 이벤트로, 정상적인 명령어 흐름을 방해합니다. 예외는 문법적으로는 문제가 없지만, 실행 시점의 상황(파일 없음, 0으로 나누기 등)으로 인해 발생합니다.

- 예시
    * ZeroDivisionError: 숫자를 0으로 나눌 때.
    * FileNotFoundError: 존재하지 않는 파일을 열려고 할 때.
    * TypeError: 올바르지 않은 타입의 피연산자를 사용할 때.
    * KeyError: 딕셔너리에 없는 키로 접근할 때.

### 예외 처리의 필요성
예외 처리를 하지 않으면 프로그램은 예외 발생 즉시 중단(Crash)됩니다. 예외 처리는 이러한 상황을 방지하고 사용자에게 유의미한 정보를 제공하며, 리소스 정리 작업(파일 닫기 등)을 보장합니다.



## 기본 구조: try...except
파이썬에서 예외 처리는 try와 except 키워드를 사용하여 이루어집니다.

```python
try:
    # 1. 예외가 발생할 가능성이 있는 코드
    value = 10 / 0
except:
    # 2. try 블록에서 예외가 발생했을 때 실행되는 코드
    print("오류가 발생했습니다.")
```
### 0 나누기 예외 처리
```python
def safe_division(numerator, denominator):
    try:
        result = numerator / denominator
        print(f"나눗셈 결과: {result}")
    except:
        # 광범위한 except 사용 예시 (일단 작동시키기 위함)
        print("경고: 0으로 나눌 수 없습니다.")

safe_division(10, 2)  # 결과: 5.0
safe_division(10, 0)  # 결과: 경고: 0으로 나눌 수 없습니다.
```


## 특정 예외 처리 (Catching Specific Exceptions)
모든 예외를 일반 except로 처리하기보다는, 발생 가능한 특정 예외를 명시하여 더욱 정밀하게 처리하는 것이 좋습니다.
### 특정 예외 명시
```python
while True:
    try:
        number = int(input("정수를 입력하세요: "))
        result = 10 / number
    except ValueError:
        print("❌ 오류: 유효한 정수 형식으로 입력해야 합니다.")
    except ZeroDivisionError:
        print("❌ 오류: 0이 아닌 숫자를 입력해야 합니다.")
    except Exception as e:
        # 그 외 모든 예외를 잡을 때 (일반적으로 가장 마지막에 위치)
        print(f"알 수 없는 오류가 발생했습니다: {e}")
    else:
        # 예외 없이 try 블록이 성공했을 때만 실행됨 (다음 섹션)
        print("✅ 성공적으로 처리되었습니다.")
        break
```
### 예외 객체 접근 (as e)
except ExceptionType as e: 구문을 사용하여 발생한 예외 객체를 변수 e에 저장하고, 예외에 대한 상세 정보를 출력할 수 있습니다.
```python
data_dict = {'apple': 5, 'banana': 10}

try:
    print(data_dict['cherry'])
except KeyError as e:
    # 예외 객체 e를 사용하여 어떤 키가 문제였는지 확인
    print(f"KeyError 발생! 딕셔너리에 '{e}' 키가 없습니다.")
```


## 선택적 블록: else와 finally
### `else` 블록 (예외가 없을 때) `try` 블록에서 예외가 발생하지 않고 성공적으로 완료되었을 경우에만 실행됩니다.
```python
try:
    x = 10
    y = 5
    result = x / y
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
else:
    # 예외가 없었을 때만 실행되는 코드
    print(f"계산 성공! 결과: {result}")
    # 이 블록에 후속 작업 코드를 배치합니다.
```
### `finally` 블록 (항상 실행되는 마무리) try 블록의 성공 여부, 예외 발생 및 처리 여부와 관계없이 항상 실행이 보장됩니다. 주로 파일 닫기, 네트워크 연결 해제 등 리소스를 정리할 때 사용됩니다.

```python
file_obj = None # 변수 초기화

try:
    file_obj = open("sample.txt", "w")
    file_obj.write("데이터 작성 중...")
    # 예외가 발생하더라도...
    # raise Exception("강제 오류 발생")
except FileNotFoundError:
    print("파일을 찾을 수 없습니다.")
finally:
    # 예외 발생 여부와 관계없이 파일 닫기 작업은 반드시 실행됨
    if file_obj:
        file_obj.close()
        print("파일 닫기 완료.")
```



## 사용자 정의 예외(raise)
`raise` 키워드
개발자가 특정 조건이 충족되지 않았을 때 의도적으로 예외를 발생시켜 프로그램의 흐름을 중단시킬 수 있습니다.
 ```python
 def set_temperature(temp):
    if not (0 <= temp <= 100):
        # ValueError 예외를 강제로 발생시키고 메시지를 전달
        raise ValueError("온도는 0에서 100 사이여야 합니다.")
    return f"온도가 {temp}°C로 설정되었습니다."

try:
    set_temperature(120)
except ValueError as e:
    print(f"⛔️ 예외 발생: {e}")
```
### 사용자 정의 예외 (Custom Exceptions)
기본 예외 클래스(Exception 등)를 상속받아 프로젝트의 특정 상황에 맞는 고유한 예외 클래스를 만들 수 있습니다.
```python
# Exception 클래스를 상속받아 새로운 예외 정의
class InsufficientFundsError(Exception):
    """잔액 부족 시 발생하는 예외"""
    def __init__(self, required, available):
        self.required = required
        self.available = available
        super().__init__(f"잔액 부족! 요청 금액: {required}, 현재 잔액: {available}")

def withdraw(amount, balance):
    if amount > balance:
        raise InsufficientFundsError(amount, balance) # 정의한 예외 발생
    return balance - amount

try:
    new_balance = withdraw(50000, 30000)
    print(f"출금 후 잔액: {new_balance}")
except InsufficientFundsError as e:
    print(f"거래 실패 상세: {e}")
```
## with 문 (Context Manager)
`finally` 블록을 사용하는 대신, `with` 문을 사용하면 파일과 같은 리소스를 자동으로 닫아주는 더 깔끔하고 권장되는 방법입니다.
```python
# try...finally를 대체하는 with 문
try:
    with open("data.txt", "r") as f:
        content = f.read()
    print("파일 읽기 성공.")
# with 블록을 벗어나는 순간 f.close()가 자동으로 호출됩니다.
except FileNotFoundError:
    print("❌ 파일을 찾을 수 없습니다.")

# with 문은 파일을 자동으로 닫아주는 __enter__와 __exit__ 메서드를 가진 객체(Context Manager)에 사용됩니다.
```
## 예외 처리 가이드라인 (Best Practices)
- ❌ 광범위한 except는 피하세요:
> except: 와 같이 특정 예외를 명시하지 않고 모든 예외를 잡는 것은, 예상치 못한 버그를 숨겨 디버깅을 어렵게 만듭니다. 항상 특정 예외를 명시하세요.
- ✅ try 블록을 최소화하세요:
> 예외가 발생할 가능성이 있는 코드만 try 블록에 넣고, 나머지는 else 블록 등으로 분리하여 코드의 가독성을 높이세요.
- ❌ 예외를 삼키지 마세요:
> 예외를 잡고 아무것도 하지 않는 것 (except SomeError: pass)은 프로그램이 조용히 실패하게 만듭니다. 최소한 콘솔에 오류 메시지를 출력하거나 로깅하세요.
- ✅ finally 또는 with로 리소스를 정리하세요:
> 파일, 네트워크 연결, 데이터베이스 연결 등은 예외 발생 여부와 관계없이 반드시 해제되도록 보장해야 합니다.
